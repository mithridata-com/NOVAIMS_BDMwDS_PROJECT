```python
data_path = r"C:\Users\dimet\Documents\GitHub\NOVAIMS_BDMwDS_PROJECT\00 Data"
```

```python
# Function to load each Excel file into its own DataFrame
def load_excel_dataframe(filename: str, sheet_name: str) -> pd.DataFrame:
    """
    Load a sheet from an Excel file into a pandas DataFrame.
    Prints the shape on success or an error message on failure.
    """
    file_path = os.path.join(data_path, filename)
    try:
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        print(f"Loaded '{sheet_name}' from '{filename}' (shape: {df.shape})")
        return df
    except Exception as e:
        print(f"Error loading '{sheet_name}' from '{filename}': {e}")
        return pd.DataFrame()
```

```python
# Load DataFrames
df1 = load_excel_dataframe("Funnel.xlsx", "funnel_data")
```

```python
df1.shape
```

```python
df1.dtypes
```

```python
# 1) split columns by dtype
num_cols = df1.select_dtypes(include="number").columns
dt_cols  = df1.select_dtypes(include="datetime64[ns]").columns
obj_cols = df1.columns.difference(num_cols.union(dt_cols))

# 2) numeric summary
stats_num = df1[num_cols].describe().T

# 3) datetime summary
stats_dt = pd.DataFrame(index=dt_cols)
stats_dt["count"] = df1[dt_cols].count()
stats_dt["min"]   = df1[dt_cols].min()
stats_dt["25%"]   = df1[dt_cols].quantile(0.25)
stats_dt["50%"]   = df1[dt_cols].median()
stats_dt["75%"]   = df1[dt_cols].quantile(0.75)
stats_dt["max"]   = df1[dt_cols].max()

# 4) object summary
stats_obj = df1[obj_cols].describe().T

# 5) stitch
stats = pd.concat([stats_num, stats_dt, stats_obj], sort=False)
stats.insert(0, "dtype", [df1[col].dtype for col in stats.index])

# 6) define your formatters
fmt_int = "{:,}"
fmt_2dp = "{:,.0f}"

def fmt_general(x):
    if pd.isna(x):
        return "—"
    if isinstance(x, pd.Timestamp):
        return x.strftime("%Y-%m-%d")
    # fallback for ints and floats
    try:
        # two‐decimal formatting (you can adjust to 1dp etc)
        return fmt_2dp.format(x)
    except Exception:
        # last‐resort: default repr
        return str(x)

# build the dict mapping *stat‐columns* to formatters
formats = {
    "mean": fmt_2dp,
    "std":  fmt_2dp,
    "count": fmt_int,
    "unique": fmt_int,
    "freq":   fmt_int,
    # for all the “edge” cols, use our general catch‐all
    **{c: fmt_general for c in ["min","25%","50%","75%","max"] if c in stats.columns}
}

# 7) style & display
styled = (
    stats.style
         .format(formats, na_rep="—")
         .background_gradient(
             subset=["count"], cmap="Greens",
             vmin=0, vmax=stats["count"].max()
         )
         .set_properties(subset=["dtype"], **{"text-align": "left"})
         .set_caption("Descriptive Statistics — df1")
         .set_table_styles([
             {"selector": "caption",
              "props": [
                  ("caption-side", "top"),
                  ("font-size", "16px"),
                  ("font-weight", "bold")
              ]},
             {"selector": "th, td",
              "props": [("border", "1px solid #ddd")]}
         ])
)
display(styled)
```

```python
# build a one-column DataFrame so you can style it
missing_tbl = (
    df1.isna().sum()               # counts
       .to_frame('Missing values') # name the column
       .sort_values('Missing values', ascending=False)  # optional: largest on top
)

# ✨ style for readability
styled1 = (
    missing_tbl.style
        .format('{:,}')                          # thousands separators
        .background_gradient(cmap="Reds")        # red heat-map for emphasis
        .set_caption('Missing-Value Summary for df1')
        .set_table_styles(
            [{'selector': 'caption',
              'props': [('font-size', '16px'),
                         ('font-weight', 'bold'),
                         ('text-align', 'left')]}]
        )
)

display(styled)      # renders as a neat HTML table you can screenshot
```
